Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

## Предобработка данных

На этапе предварительного анализа были замечены следующие проблемы:
- В датафрейме contract присутствуют пропуски в столбце TotalCharges, их следует обработать.
- В contract столбец BeginDate следует перевести в тип datetime.
- Названия колонок во всех датафреймах в "верблюжем регистре", это может создать некоторые трудности для дальнейшей работы, но это некритично - оставим как есть.

### Обработка пропусков

Обработаем пропуски.

```python
#вывод пропусков
display(contract.query('TotalCharges.str.isspace()'))
# индексы строк с пропусками
idx = contract.query('TotalCharges.str.isspace()').index

print('Кол-во пропусков: ', len(contract.query('TotalCharges.str.isspace()')))
```

Выгрузка данных была произведена 1 февраля 2020 года. Все вышеперечисленные абоненты заключили договор также 1 февраля 2020 года, то есть они пользовались услугами оператора менее месяца. Заполняем пропуски значением из столбца месячной платы.

```python
# заполнение пропусков, значением из столбца месячной платы
contract['TotalCharges'] =\
    contract['TotalCharges'].where(cond=(~contract['TotalCharges'].str.isspace()), other=contract['MonthlyCharges'])
```

```python
# проверка на заполнение пропусков
contract.iloc[idx]
```

```python
# преобразование к float
contract['TotalCharges'] = contract['TotalCharges'].astype('float')

# проверка
print('Новый тип данных столбца TotalCharges:', contract['TotalCharges'].dtype)
print('Кол-во пропусков:', contract['TotalCharges'].isna().sum())
```

Пропуски заполнены, тип данных изменен.

### Изменение типа данных

### Создание новых признаков

Добавим новые столбцы:
- кол-во дней пользования услугами
- метка о том, пользуется ли абонент телефоном/интернетом или всем сразу.
- метка о том, ушел ли клиент.

Новый столбец с кол-вом дней пользования услугами totalDays.

```python
# формирование Series с датой бэкапа данных
backup_date = pd.Series(pd.to_datetime(arg='2020-02-01', format='%Y-%m-%d'), index=data.loc[data['EndDate'] == 'No'].index)

# заполнение нового столбца для абонентов, которые продолжают пользоваться услугами
data.loc[data['EndDate'] == 'No', 'totalDays'] =\
    (backup_date-data['BeginDate']).dt.days

# заполнение нового столбца для абонентов, которые прекратили пользоваться услугами
data.loc[data['EndDate'] != 'No', 'totalDays'] =\
    (pd.to_datetime(data.loc[data['EndDate'] != 'No', 'EndDate'], format='%Y-%m-%d %H:%M:%S') - data['BeginDate']).dt.days

display(data.head())
print('Пропуски в столбце totalDays:', data['totalDays'].isna().sum())
```

Создадим столбец с меткой о том, пользуется ли абонент телефоном/интернетом или всем сразу. Назовем `domain`.  
Следующие метки будут означать:
- пользователь только интернета - 0;
- только телефон - 1;
- все сразу - 2.

```python
data.loc[data['customerID'].isin(internet['customerID']), 'domain'] = 0
data.loc[data['customerID'].isin(phone['customerID']), 'domain'] = 1
data.loc[data['customerID'].isin(internet['customerID']) & data['customerID'].isin(phone['customerID']), 'domain'] = 2

display(data.head())
print('Пропуски в столбце domain:', data['domain'].isna().sum())
```

Столбец с меткой о том, ушел ли клиент назовем `churn`. 0 - будет соответствовать действующим пользователям, 1 - ушедшим.

```python
data.loc[data['EndDate'] == 'No', 'churn'] = 0 
data.loc[data['EndDate'] != 'No', 'churn'] = 1

display(data.head())
print('Пропуски в столбце churn:', data['churn'].isna().sum())
```

Выясним, можно ли сказать, что средние значения месячной платы для действующих и ушедших пользователей отличаются. Воспользуемся статистической проверкой гипотез.  

H0: средние действующих и ушедших клиентов равны.  
H1: среднее действующих < среднего ушедших клиентов.  
Уровень значимости alpha определим как 0.05.
Для расчета p-value используем t-test для двух независимых выборок с разной дисперсией.

```python
current_customers = data.loc[data['churn'] == 0]['MonthlyCharges']
churn_customers = data.loc[data['churn'] == 1]['MonthlyCharges']
alpha=0.05

result = stats.ttest_ind(current_customers,
                                  churn_customers, 
                                  equal_var=False, 
                                  alternative='less')

print('Значение pvalue =', result.pvalue)
if result.pvalue < alpha:
    print('Принимаем альтернативную гипотезу')
else:
    print('Не удалось отклонить нулевую гипотезу')
```

У действующих и ушедших клиентов средние значения месячной платы - разные, причем у действующих клиентов среднее меньше, чем у ушедших.

Выясним, можно ли сказать, что средние значения итоговой платы для действующих и ушедших пользователей отличаются. Воспользуемся статистической проверкой гипотез.  

H0: средние действующих и ушедших клиентов равны.  
H1: среднее действующих > среднего ушедших клиентов.  
Уровень значимости alpha определим как 0.05.
Для расчета p-value используем U-критерий Манна-Уитни, т.к. в выборке ушедших клиентов есть выбросы.

```python
current_customers = data.loc[data['churn'] == 0]['TotalCharges']
churn_customers = data.loc[data['churn'] == 1]['TotalCharges']
alpha=0.05

result = stats.mannwhitneyu(current_customers,
                            churn_customers,
                            alternative='greater')

print('Значение pvalue =', result.pvalue)
if result.pvalue < alpha:
    print('Принимаем альтернативную гипотезу')
else:
    print('Не удалось отклонить нулевую гипотезу')
```

### Создание пайплайна

Воспользуемся пайплайном. Это инструмент, который облегчает предобработку и трансформацию данных, помещая их в один поток.
Разделим признаки на категориальные, бинарные и числовые. Делается это для того, чтобы была возможность применять к ним разные подходы в обработке данных.  

Для числовых признаков используем стандартизацию (StandardScaler) и удаление сильно коррелирующих призанков (DropCorrelatedFeatures), они необходимы для линейных моделей для адекватной отработки алгоритма. Остальных модели в данных процедурах не нуждаются.  
Категориальных данные закодируем порядковым кодировщиком (OrdinalEncoder).   
Для бинарных признаков также воспользуемся кодировщиком, но уже OneHotEncoder.  
Целевой признак оставляем без изменения.

Отбор признаков в пайплайне осуществляется с помощью DropCorrelatedFeatures и SelectKBest. 

```python
#деление признаков на категориальные, бинарные и численные
cat_columns = ['Type', 'PaymentMethod', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 
              'StreamingTV', 'StreamingMovies', 'MultipleLines', 'domain']
bin_columns = ['PaperlessBilling', 'SeniorCitizen', 'Partner', 'Dependents', 'gender']
num_columns = ['MonthlyCharges', 'totalDays', 'TotalCharges']

# пайплайн для числовых признаков (создается отдельно, чтобы в будущем была возможность использовать признаки без скейлера)
num_pipeline = Pipeline([('scaler', StandardScaler()),
                         ('drop_correlations', DropCorrelatedFeatures())])

# ColumnTransformer объединяет результат трансформаций для разных типов колонок
columns_pipeline=ColumnTransformer([('num_cols', num_pipeline, num_columns),
                                    ('category_encoder', OrdinalEncoder(), cat_columns),
                                    ('binary_encoder', OneHotEncoder(drop='first'), bin_columns)],
                                   remainder='passthrough')

# оборачивание ColumnTransformer в пайплайн
pipeline = make_pipeline(columns_pipeline)
pipeline
```

## Обучение моделей

Представленные модели интерпретируемы, не требуют большого объема вычислительных ресурсов и эффективны для небольшого объема данных.

Подбор гиперпараметров будем производить с помощью RandomizedSearchCV.

Лучшую модель проверим на тестовой выборке и дополнительно измерим метрику Accuracy.

Борьбу с дисбалансом классов попробуем решить с помощью встроенного в модели параметра `class_weight='balanced'`. Если данный способ не поможет, можно прибегнуть к апсемплингу, например, с помощью SMOTETomek.

### Предобработка данных

Предварительно данные были проверены на пропуски, дубликаты и выбросы. В столбце `TotalCharges` нашлось 11 пропущенных значений - пустых строк. Пропуски заполнены значением из столбца `MonthlyCharges`. Далее, была произведена коррекция типов данных: для числовых данных был присвоен тип int, для категориальных - category.  

Созданы 3 синтетических признака. 
- `churn` - метка об уходе клиента: 0 - не ушел, 1 - ушел; создан на факте наличия даты окончания договора в `EndDate`;
- `domain` - категория пользователя услугам: 0 - интернет, 1 - телефон, 2 - интернет + телефон;
- `totalDays` - длительность контракта в днях; создан на основе разницы даты окончания и начала договора: `EndDate` - `BeginDate`. Если у клиента не проставлено значение `EndDate`, то вместо неё использовалась дата выгрузки данных `2020-02-01`.

Удаленные признаки:
- `customerID` - идентификатор клиента; для модели информации не несет,
- `BeginDate`- дата начала действия договора; неинформативный признак,
- `EndDate` - дата окончания договора; признак вызывает утечку данных, т.к. из него формируется целевой признак.  

Данные были разделены на обучающую и тестовую выборки в соотношении 75:25.
Затем используемые признаки были разбиты на 3 группы: числовые, категориальные и бинарные. Такой подход обусловлен тем, что разные типы данных требуют конкретных методов предобработки.  

К числовым признакам применены следующие процедуры: стандартизация (StandardScaler) - приведение к общей числовой шкале; удаление сильно коррелирующих признаков (DropCorrelatedFeatures). Примечание: не все рассматриваемые модели нуждаются в стандартизации числовых признаков, это критично только для модели логистической регрессии (LogisticRegression) при её использовании без регуляризации - метода предотвращения переобучения модели, когда она подстраивается под обучающую выборку и неспособна "предсказывать" на других данных. 
Пайплайн предоставляет возможность отключить стандартизацию числовых признаков для моделей, в которых в ней нет необходимости.

Бинарные признаки были закодированы с помощью OneHotEncoder. Такое преобразование создает дополнительное количество столбцов-признаков, соответствующее числу уникальных значений кодируемого признака. В то же время, значения кодируемого признака преобразуются в 0 и 1, и проставляются в соответствующие столбцы.

Категориальные признаки кодируются порядковым кодировщиком (OrdinalEncoder): каждой категории ставится в соответствие целое число.

Предобработка бинарных и категориальных признаков требуется по той причине, что модели (в общем случае) не работают со строковыми переменными, им нужна числовая интерпретация.

Целевой признак оставляем без изменения.

Вышеперечисленные преобразования удалось сделать для каждого типа признака отдельно благодаря ColumnTransformer. Дальше, с его помощью, предобработанные признаки снова объединяются в единое пространство, с которым уже могут работать модели.

Последним шагом перед поступлением данных в модель является отбор наиболее релевантных признаков с помощью SelectKBest. Данный метод проставляет каждому признаку оценку на основе статистических тестов.
